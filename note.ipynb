```python
bad = ['Default','Paid Off','Charged Off','Delinquent']
good = ['Settled','Paid Off']
tbd = ['Current']
```

```python
bad_account = df.query("loan_status in @bad")
good_account = df.query("loan_status in @good")
tbd_account = df.query("loan_status in @tbd")
```

```python
bad_account['target'] = 1
good_account['target'] = 0
tbd_account['target'] = 0
```

```python
tbd_account['ratio']  = tbd_account['num_successful_payments'] / tbd_account['loan_term_months'] 
```

```python
tbd_account['ratio'] = tbd_account['num_successful_payments'] / tbd_account['loan_term_months'] 
print(tbd_account[tbd_account['ratio']>0.6].shape)
print(tbd_account[tbd_account['ratio']>0.6].shape[0]/tbd_account.shape[0])
tbd_account_filtered = tbd_account[tbd_account['ratio']>0.6].drop(columns='ratio')
tbd_account_filtered
```

```python
final_v2 = pd.concat([bad_account, good_account, tbd_account_filtered], axis=0, ignore_index=True)
final_v2['autopay_enrolled'] = final_v2['autopay_enrolled'].replace({True: 'yes', False: 'no'})
```

```python
final_v2[['annual_income','loan_amount','loan_term_months','credit_score','autopay_enrolled','target']].to_csv("loan_ready_ml_v4.csv", index = False)
```
